We need selenium to extract dynamic data rendered through Javascript. 

Download chromedriver.exe from https://sites.google.com/a/chromium.org/chromedriver/downloads (chromedriver version must match with Google Chrome version)

```python
# python -m pip install selenium
from selenium import webdriver
import time
from bs4 import BeautifulSoup           # Library to parse data from webpages
```

```python
#some of the top companies looking for data scientists in the US (according to Glassdoor)
#Adobe, Amazon, Google, McKinzie, NVDIA
url = 'https://www.glassdoor.com/Reviews/Adobe-Reviews-E1090.htm'
driver = webdriver.Chrome('c:/Users/TV/Downloads/chromedriver_win32/chromedriver')
driver.get(url)
```

```python
time.sleep(1)                              # Give Javascript time to render
soup = BeautifulSoup(driver.page_source)
```

```python
print(soup.find(class_="empReview").prettify())
```

```python
#create table to store results
import pandas as pd
df = pd.DataFrame()
```

```python
#create function for adding tables
def add_column(query,column_name):
    col = []
    if type(query) is list:
        for i in soup.find_all(attrs={query[0]:query[1]}):
            col.append(i.get_text())
    else:
        for i in soup.find_all(class_=query):
            col.append(i.get_text())
    df[column_name] = col 
```

```python
#create table to store results
import pandas as pd
df = pd.DataFrame()

#add special column id
r_id = []
for i in soup.find_all(class_='empReview'):
    r_id.append(i.get('id'))
df['ID'] = r_id

#use function for all other columns
add_column('ratingNumber','Rating')
add_column('reviewLink','General')
add_column(["data-test","pros"],'Pros')
add_column(["data-test","cons"],'Cons')
add_column('eg4psks0','Status')
add_column('authorLocation','Location')
add_column('authorJobTitle','Date/Title')
df
```

```python
#find total number of webpages to loop through
footer = soup.find(class_='paginationFooter').get_text()
num = footer.split()
per_page = int(num[3])-int(num[1])+1
total = int(num[5].replace(',',''))

print(footer)
print("per_page: "+str(per_page))
print("total: "+str(total))

import math
pages = math.ceil(total/per_page)
pages
print("pages: {}".format(pages))
```
